```python
import pandas as pd
import matplotlib.pyplot as plt
from wordcloud import WordCloud
```

```python
df = pd.read_csv("C:\\Users\\dudde\\Downloads\\tweets_large.csv")

print(df.head())
```

```python
text = " ".join(df['text'])
```

```python
wordcloud = WordCloud(
    width=800,
    height=400,
    background_color='white'
).generate(text)
```

```python
plt.figure(figsize=(10,5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title("Word Cloud of Tweets")
plt.show()
```

```python
positive_text = " ".join(df[df['sentiment'] == 'Positive']['text'])

wordcloud_pos = WordCloud(background_color='white').generate(positive_text)

plt.imshow(wordcloud_pos)
plt.axis('off')
plt.title("Positive Tweets Word Cloud")
plt.show()
```

```python
negative_text = " ".join(df[df['sentiment'] == 'Negative']['text'])

wordcloud_neg = WordCloud(background_color='black').generate(negative_text)

plt.imshow(wordcloud_neg)
plt.axis('off')
plt.title("Negative Tweets Word Cloud")
plt.show()
```
